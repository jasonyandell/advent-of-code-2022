```python
sample = """498,4 -> 498,6 -> 496,6
503,4 -> 502,4 -> 502,9 -> 494,9"""

def add_line(points, a, b):
    x_start,y_start = a
    x_end,y_end = b
    to_range = lambda a,b: range(min(a,b),max(a,b)+1)
    for x in to_range(x_start, x_end):
        for y in to_range(y_start, y_end):
            points.add((x,y))

def parse(input:str):
    points = set()
    def get_coord(c:str):
        cs = c.split(',')
        return (int(cs[0]),int(cs[1]))
    lines = input.splitlines()
    coord_str_lines = [line.split(' -> ') for line in lines]
    for coord_str_line in coord_str_lines:
        start = get_coord(coord_str_line[0])
        for c in coord_str_line[1:]:
            next = get_coord(c)
            add_line(points, start,next)
            start = next
    return points


def print_points(points):
    left = min(x for x,_ in points)
    right = max(x for x,_ in points)
    top = max(y for _,y in points)

    print(left, right, 0, top)
    for y in range(0,top+1):
        line = ""
        for x in range(left-1,right+2):
            if (x,y) in points: line += '#'
            else: line += '.'
        print(line)
    print()

def add(p1,p2): return (p1[0]+p2[0],p1[1]+p2[1]) #tuple(map(lambda a,b:a+b, p1, p2))

def drop(points, top):
    point = (500,0)
    if point in points: return True
    while True:
        if (point[1] > top): break

        # try down, down left, down right
        down = add(point, (0,1))
        down_left = add(point, (-1,1))
        down_right = add(point, (1,1))

        # learned for..else from reddit
        for p in down, down_left, down_right:
            if p not in points:
                point = p
                break
        else: break
    if point[1] < top:
        points.add(point)
        return True
    return False

def part1(points):
    start = (500,0)
    top = max(y for _,y in points) + 3 # + 3 to make room for the floor in part 2
    grains = 0
    while start not in points and drop(points, top): grains += 1
    return grains

def part2(points):
    # put a big line at top+2
    left = min(x for x,_ in points)
    right = max(x for x,_ in points)
    top = max(y for _,y in points)    
    add_line(points, (left-1000,top+2), (right+1000, top+2))
    return part1(points)

print_points(parse(sample))

assert 24 == part1(parse(sample))
assert 93 == part2(parse(sample))
```

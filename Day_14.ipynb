```python
sample = """498,4 -> 498,6 -> 496,6
503,4 -> 502,4 -> 502,9 -> 494,9"""

def draw(points, a,b):
    ar,ac = a
    br, bc = b
    if ar > br: br, ar = ar, br
    if ac > bc: bc, ac = ac, bc 
    for r in range(ar, br+1):
        for c in range(ac, bc+1):
            points.add((r,c))

def parse(input:str):
    points = set()

    def get_coord(c:str):
        cs = c.split(',')
        return (int(cs[0]),int(cs[1]))

    lines = input.splitlines()
    coord_str_lines = [line.split(' -> ') for line in lines]
    for coord_str_line in coord_str_lines:
        start = get_coord(coord_str_line[0])
        for c in coord_str_line[1:]:
            next = get_coord(c)
            draw(points, start,next)
            start = next
    
    return points


def print_points(points, left, right, bottom, top):
    print(left, right, bottom, top)

    for y in range(0,top+1):
        line = ""
        for x in range(left,right+1):
            if (x,y) in points: line += '#'
            else: line += '.'
        print(line)
    print()


def add(p1,p2): return tuple(map(lambda a,b:a+b, p1, p2))

def drop_down(points, start, top):
    point = start
    if point in points: return point
    while True:
        # try down, down left, down right
        down = add(point, (0,1))
        down_left = add(point, (-1,1))
        down_right = add(point, (1,1))
        if (down[1] > top): break
        if down not in points: point = down
        elif down_left not in points: point = down_left
        elif down_right not in points: point = down_right
        else: break

    return point

def drop(points, start, top):
    point = drop_down(points, start, top)
    if point[1] < top:
        #print(point)
        points.add(point)
        return True
    print("eaten by a grue")
    return False

def part1(data):
    points = parse(data)
    left = min(x for x,_ in points)
    right = max(x for x,_ in points)
    bottom = min(y for _,y in points)
    top = max(y for _,y in points)

    print(top)
    #print_points(points, left, right, bottom, top)

    i = 0
    start = (500,0)
    while i < 3000 and drop(points, start, top):
        i += 1
        #print_points(points, top)
    return i


def part2(data):
    print('part 2')
    points = parse(data)
    left = min(x for x,_ in points)
    right = max(x for x,_ in points)
    bottom = min(y for _,y in points)
    top = max(y for _,y in points)

    draw(points, (left-1000,top+2), (right+1000, top+2))
    print_points(points, left, right, bottom, top+2)

    i = 0
    start = (500,0)
    while i < 100000 and drop(points, start, top+3):
        i += 1
        if start in points: break
        #print_points(points, left, right, bottom, top)
    return i

assert 24 == part1(sample)
assert 93 == part2(sample)
```
